## Zadanie nr 1

Wiedząc, że spaCy nie radzi sobie z tokenizacją nadmiarowych białych znaków, zaproponuj jak można wstępnie przygotować tekst do przetworzenia.

````
text ="""Ala  ma kota.
Kot   jest   biały."""

for t in nlp(text):
  print(f"[{t}]")
````

Oczekiwany wynik:

```
[Ala]
[ma]
[kota]
[.]
[Kot]
[jest]
[biały]
[.]
```

### Rozwiązanie

````
import re

text_fixed = re.sub("\s+", " ", text)

for t in nlp(text_fixed):
  print(f"[{t}]")
````

````
for line in re.split("\s*\n+\s*", text):
  if line.strip() == "":
    continue

  text_line = re.sub("\s+", " ", line)

  for t in nlp(text_line):
    print(f"[{t}]")
  print("--")
````

## Zadanie nr 2

Wypisz 10 najczęstszych bigramów lematów składających się ze słów o długości nie mniejszej niż 4 znaki. Dla najczęstszego bigramu wypisz częstość wszystkich form tekstowych.

Np. załóżmy że najczęstszym bigramem lematów będzie "zielony drzewo", a jego formy tekstowe to:
* zielonego drzewa 10
* zielonym drzewem 5

### Rozwiązanie

````
from nltk.probability import FreqDist
from nltk import bigrams
import re

# Pomocnicze wyrażenia regularne
words_regex = re.compile(".{4,} .{4,}")
spaces_regex = re.compile("\s+")

# Do zliczenia wszystkich bigramów lematów
fq = FreqDist()

# Słownik lemat => FreqDist dla form tekstowych
bigram_forms = {}

for text in texts:
  doc = nlp(spaces_regex.sub(" ", text), disable=["ner"])

  for sentence in doc.sents:
    tokens = doc[sentence.start:sentence.end]

    for ngram in bigrams(tokens):
      ngram_lemma = " ".join([t.lemma_ for t in ngram])
            
      if not words_regex.fullmatch(ngram_lemma):
        continue

      fq.update([ngram_lemma])
      
      ngram_orth = " ".join([t.text for t in ngram])
      bigram_forms.setdefault(ngram_lemma, FreqDist()).update([ngram_orth])

print(*fq.most_common(10), sep="\n")
````

````
top_bigram = fq.most_common(1)[0][0]
print(top_bigram)
print("-"*30)
print(*bigram_forms.get(top_bigram).most_common(10), sep="\n")
````

## Zadanie nr 3

Znajdź w zbiorze parę dokumentów najbardziej podobnych do siebie. Dla pierwszego dokumentu z pary wypisz wszystkie zdania z przypisaniem następującej informacji:
* `[=]` — zdanie występuje w drugim dokumencie,
* `[~]` — zdanie jest podobne do jednego ze zdań w drugim dokumencie. Należy wypisać jakie to zdanie. Zdania są do siebie podobne, jeżeli odległość edycyjna między jednym i drugim zdaniem wynosi mniej niż 12 zgodnie z miarą Damerau-Levenhstein.
* `[+]` — zdanie nie występuje w drugim dokumencie.

**Przykład**

Załóżmy, że te dwa dokumenty są do siebie najbardziej podobne:

```
doc1 = "Ala ma kota. Kot jest czarny. Kot nazywa się mruczek."
doc2 = "Ala ma kota. Kot jest prawie czarny."
```

*Wynik:*

```
[=] Ala ma kota.
[~] Kot jest czarny.
    Kot jest prawie czarny.
[+] Kot nazywa się mruczek.
```



### Rozwiązanie

````
from sklearn.feature_extraction.text import TfidfVectorizer

# wektoryzacja dokumentów
vectorizer = TfidfVectorizer(ngram_range=(1,2), min_df=2)
X = vectorizer.fit_transform(texts)

# liczba cech
print(len(vectorizer.get_feature_names()))
````

````
from sklearn.metrics.pairwise import cosine_similarity

# macierz podobieństwa
sims = cosine_similarity(X, X)
````

```

```

```
[0]  -   a   b   c
[1]  -   -   d   e
[2]  -   -   -   f
[3]  -   -   -   -
```

```
[(a,0,1), (b,0,2), (c,0,3), ...]
```

````
# lista krotek (podobieństwo, doc1, doc2)
sims_tuple = []
for n in range(len(texts)):
  for m in range(n+1, len(texts)):
    sims_tuple.append((sims[n,m], n, m))
````

````
# pięć najbardziej podobnych par dokumentów
print(sorted(sims_tuple, reverse=True)[:5])
````

````
# nazwa pliku i url dla najbardziej podobnych dokumentów
top = sorted(sims_tuple, reverse=True)[0]
print(top[0])
print(files[top[1]], urls[top[1]], sep="\n")
print(files[top[2]], urls[top[2]], sep="\n")
````

````
# podział dokumentów na zdania
text1 = texts[top[1]]
text2 = texts[top[2]]

doc1 = nlp(text1)
doc2 = nlp(text2)

sentences1 = [s.text.strip() for s in doc1.sents]
sentences2 = [s.text.strip() for s in doc2.sents]

print("Dokument A")
print("-"*20)
print(*sentences1, sep="\n")
print("")
print("Dokument B")
print("-"*20)
print(*sentences2, sep="\n")
````

````
def find_most_similar(s, sentences, threshold=12):
  '''
  Pomocnicza funkcja do znalezienia najbardziej podobnego zdania.
  '''
  sims = [(jellyfish.damerau_levenshtein_distance(s, sent), sent) for sent in sentences]
  sims = [(dist, sent) for dist, sent in sims if dist < threshold]
  if len(sims) == 0:
    return None
  else:
    return sorted(sims)[-1][1]
````

````
# analiza zdań dla pierwszego dokumentu z pary
for s in sentences1:
  if s in sentences2:
    print(f"[=] {s}")
    continue

  most_similar = find_most_similar(s, sentences2)
  if most_similar == None:
    print(f"[+] {s}")
    continue

  print(f"[~] {s}")
  print(f"    {most_similar}")
````
